```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
frame=cv2.imread("LenaSoderberg.jpg")
```

```python
plotSize=[3,2]
cv2bgr2rgb=lambda x: cv2.cvtColor(x,cv2.COLOR_BGR2RGB)
cv2bgr2gray=lambda x: cv2.cvtColor(x,cv2.COLOR_BGR2GRAY)
axIter=iter(range(1,plotSize[0]*plotSize[1]+1))
plt.figure(figsize=(5*plotSize[1],5*plotSize[0]))
plt.subplot(*plotSize,next(axIter))
plt.imshow(cv2bgr2rgb(frame)) # matplotlib use RGB and CV2 use BGR
plt.title("Original")

plt.subplot(*plotSize,next(axIter))
plt.imshow(cv2bgr2gray(frame),cmap="gray") # matplotlib use RGB and CV2 use BGR
plt.title("Grayscale")

plt.subplot(*plotSize,next(axIter))
plt.imshow(cv2bgr2rgb(cv2.GaussianBlur(frame,(101,101),10))) # matplotlib use RGB and CV2 use BGR
plt.title("GaussianBlur")

plt.subplot(*plotSize,next(axIter))
gx=cv2.Sobel(np.float32(cv2bgr2gray(frame)),cv2.CV_32F,1,0,3) # 32F for 32-bit float, dtype of the original image is uint8
plt.imshow(cv2bgr2rgb(gx)) 
plt.title("Sobel x")

plt.subplot(*plotSize,next(axIter))
gy=cv2.Sobel(np.float32(cv2bgr2gray(frame)),cv2.CV_32F,0,1,3) # 32F for 32-bit float, dtype of the original image is uint8
plt.imshow(cv2bgr2rgb(gy)) 
plt.title("Sobel y")
# plt.imshow(cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY),cmap="gray") # matplotlib use RGB and CV2 use BGR
# plt.title("Graysc/ale")
# cv2.imshow("Test",frame)
```
